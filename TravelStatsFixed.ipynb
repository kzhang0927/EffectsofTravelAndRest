# Scrape Data From BballReference

```python
#BeautifulSoup
for year in range(beg_year, end_year + 1):
    for team in list_teams:
        try:
            r = get(f'https://widgets.sports-reference.com/wg.fcgi?css=1&site=bbr&url=%2Fteams%2F{team}%2F{year}_games.html&div=div_games')
            if r.status_code==200:
                soup = BeautifulSoup(r.content, 'html.parser')
                table = soup.find('table')
            df2 = pd.read_html(str(table))[0]
            df2 = df2.assign(Year = year,Team = team)
            df = pd.concat([df,df2])
            sleep(3)
        except:
            pass
df
```

```python
#drop columns
df = df.drop(columns = ['Unnamed: 7','Unnamed: 8','G','Unnamed: 3','Unnamed: 4','W','L','Streak','Notes'])
```

```python
#rename columns
df = df.rename(columns={'Tm':'TeamPts','Opp':'OppPts','Unnamed: 5':'Away?'})
```

```python
#bring in team name
df['TeamName'] = df['Team'].map(team_mapping)
df['TeamName'].unique()
```

# Bring in current location 

```python
#Mapping function
def location_mapping(row):
    if row['Away?'] == '@':
        x = Geolocation[row['Opponent']]
    else:
        x = Geolocation[row['TeamName']]
    return x
```

```python
#clean up dtypes
df['Opponent'].astype(str)
df['TeamName'].astype(str)
```

```python
df['Location'] = df.apply(location_mapping, axis=1)
```

```python
df = df[df['Date'] != 'Date']
```

```python
df['PreviousLocation'] = df['Location'].shift()
```

# Calculate Distance

```python
df[['StartLat', 'StartLon']] = df['PreviousLocation'].apply(pd.Series)
df[['EndLat', 'EndLon']] = df['Location'].apply(pd.Series)
```

```python
def haversine(row):
    lon1 = row['StartLon']
    lat1 = row['StartLat']
    lon2 = row['EndLon']
    lat2 = row['EndLat']
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * arcsin(sqrt(a)) 
    km = 6367 * c
    return km
```

```python
df['DistanceFromPrevious'] = df.apply(lambda row: haversine(row), axis=1)
df
```

```python
#remove temp columns
df = df.drop(columns = ['StartLat','StartLon','EndLat','EndLon'])
```

# Converting to datetime

```python
#Temporary for datetime
df[['DOW', 'Date']] = df['Date'].str.split(' ', 1, expand=True)
df[['Month','Rest']] = df['Date'].str.split(' ', 1, expand=True)
df[['Day','Year']] = df['Rest'].str.split(',', 1, expand=True)
df
```

```python
#Fix naming error
df = df.drop(columns = ['Rest','DOW'])
```

```python
Month_conversion = {'Oct':'10','Nov':'11','Dec':'12','Jan':'01','Feb':'02','Mar':'03','Apr':'04','May':'05','Jun':'06'}
df['Mon']=df['Month'].map(Month_conversion)
```

```python
df['Year'] = df['Year'].astype('int32')
df['Day'] = df['Day'].astype('int32')
df['Mon'] = df['Mon'].astype('int32')
```

```python
df['Datetime'] = pd.to_datetime(df.Year*10000+df.Mon*100+df.Day,format='%Y%m%d')
```

```python
#drop temp columns
df = df.drop(columns = ['Month','Day','Mon','Year','Date'])
```

```python
#convert hours to 24 hr format
df['Start (ET)'] = pd.to_datetime(df['Start (ET)']).dt.strftime('%H:%M:%S')
```

```python
# merge date and hours
df['Datetime'] +=  pd.to_timedelta(df['Start (ET)'])
```

# Find Difference in Time

```python
#1 day = 86400 seconds
df['DaysBetweenGames'] = (df['Datetime'] - df['Datetime'].shift()).dt.total_seconds() / 86400
```

# Final Cleanup

```python
#Ensure teams don't reference one another
df.loc[(df['TeamName'] != df['TeamName'].shift()),'PreviousLocation']=''
df = df[df['PreviousLocation'] != '']
```

```python
#change datatype
df['TeamPts'] = df['TeamPts'].astype('float64')
df['OppPts'] = df['OppPts'].astype('float64')
```
